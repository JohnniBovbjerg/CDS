```python
import pandas as pd
#Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

my_bday = 1989

while True:
    for df in rawdata:
        if 'Released' in df.columns:
            matches = df[df['Released'] == my_bday]
        
            if not matches.empty:
                print(f"YES! I found a match in {my_bday}: thanks for waiting")
                for index, row in matches.iterrows():
                    print(f"Artist: {row['Artist']}, Single: {row['Single']}, this one rocks!")
                break  
    else:
        print(f"No entries for {my_bday}. Checking {my_bday + 1}...")
        my_bday += 1
        continue  
    break  
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
import pandas as pd
#Make a single dataframe that combines the space missions from the 1950's to the 2020's

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
df = rawdata[1]
df
```

```python
#Write a script that returns the year with the most launches
df['Year'] = pd.to_datetime(df['Launch date'], errors='coerce').dt.year

launch_counts = df['Year'].value_counts()

most_launches_year = launch_counts.idxmax()
most_launches_count = launch_counts.max()

print(f"The year with the most launches is {most_launches_year} with {most_launches_count} launches.")
```

```python
#Write a script that returns the year with the most launches 
df['Year'] = pd.to_datetime(df['Launch date'], errors='coerce').dt.year

most_launches_year = df['Year'].mode()[0]

print(f"The year with the most launches is {most_launches_year} HAHA COMPUTERS")
```

```python
#Write a script that returns the most common month for launches
df['Month'] = pd.to_datetime(df['Launch date'], errors='coerce').dt.month_name()

most_common_month = df['Month'].mode()[0]

print(f"The most common month for launches is {most_common_month}")
```

```python
#Write a script that returns the most common month for launches
df['Month'] = pd.to_datetime(df['Launch date'], errors='coerce').dt.month_name()

month_counts = df['Month'].value_counts()

most_common_month = month_counts.idxmax()
most_common_month_count = month_counts.max()

print(f"The most common month for launches is {most_common_month} with {most_common_month_count} launches THIS TIME IT MAKES SENSE")
```

```python
#Write a script that ranks the months from most launches to fewest launches
df['Month'] = pd.to_datetime(df['Launch date'], errors='coerce').dt.month_name()

month_counts = df['Month'].value_counts()

ranked_months = month_counts.sort_values(ascending=False)

print("Ranked months from most to fewest launches:")
print(ranked_months)
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
import pandas as pd
#Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
combined_df = pd.concat(tables, ignore_index=True)
combined_df.columns = combined_df.columns.str.strip().str.lower().str.replace(' ', '_')

supervillains_df = combined_df.dropna(subset=['character_/_team', 'year_debuted'])[['character_/_team', 'year_debuted']]

print(supervillains_df.to_string(index=False))
```

```python
result = supervillains_df[supervillains_df['character_/_team'].str.contains('Doctor Doom')]

print(f"{result}, He is my favorite villain, and surprisingly enough, quite old. No wonder he is so grumpy!")
```

```python
import pandas as pd
#Write a script that ranks each decade in terms of how many supervillains debuted in that decade

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
combined_df = pd.concat(tables, ignore_index=True)
combined_df.columns = combined_df.columns.str.strip().str.lower().str.replace(' ', '_')

combined_df['year_only'] = combined_df['year_debuted'].str.extract(r'(\d{4})').dropna().astype(int)

combined_df['decade'] = (combined_df['year_only'] // 10) * 10 

ranked_decades = combined_df['decade'].value_counts().sort_values(ascending=False)

print(ranked_decades)
```

```python
import pandas as pd
#Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
combined_df = pd.concat(tables, ignore_index=True)
combined_df.columns = combined_df.columns.str.strip().str.lower().str.replace(' ', '_')

company_counts = combined_df['company'].value_counts()

ranked_companies_df = pd.DataFrame({'Company': company_counts.index, 'Supervillains': company_counts.values})

print(ranked_companies_df.to_string(index=False))
print(f"Note: I think this is a pretty price table,not Ikea but still pretty nice")
```
