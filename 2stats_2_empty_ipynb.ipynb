First let's make frequency distributions for the women and men seperately.

You can make some quick and dirty separate plots using a basic function and indexing...

```r
str(heights)
```

... though it would be better if we could see all the data together on the same histogram

```r
hist(heights$height[heights$sex=="Male"])
hist(heights$height[heights$sex=="Female"])
```

shows frequency - how often a score occurs. 

```r
ggplot(heights,
       aes(x=height,
           fill=sex)) +
    geom_histogram(position='identity', 
                   alpha=0.5)
```

```r
stats_table <- heights %>%
     group_by(sex) %>%
     summarise(mean_height = mean(height),
               median_height = median(height),
               sd_height = sd(height))
```

```r
stats_table
```

Shapiro text testing how close the data is the normal distribution - W should be close to 1, and p-value should be small. which they are in this case. 

```r
shapiro.test(heights$height[heights$sex=='Male'])
```

```r
shapiro.test(heights$height[heights$sex=='Female'])
```

Let's look at the summary statistics - you can actually make them in 1 table!

Lets try visualising the data in terms of probability distributions

```r
ggplot(heights, 
      aes(x = height,
         fill = sex)) +
   geom_density(alpha=0.5)
```

Now we are ready to run a hypothesis test. Are men taller than women? What sort of t test should we do?

```r
t.test(height ~ sex, data=heights)
```

~ = relationship. 
We are determining that the two groups are not the same - falsifying, (not that they are different).

Symbol Breakdown
<-: Assignment operator. Saves the result on the right into the variable on the left, here saving it as stats_table.

%>%: Pipe operator. Passes the result of the previous line as input to the next line. In simple terms, it connects commands in a sequence.

filter(): Function that filters or selects rows based on a condition.

Species %in% c("setosa", "versicolor"): Keeps only rows where Species is either "setosa" or "versicolor".
%in%: Checks if each value on the left exists in the set on the right.

Species %in% c("setosa", "versicolor") returns TRUE for rows where Species is "setosa" or "versicolor", allowing filter() to keep only those rows.
group_by(): Groups data by one or more columns, here by Species. After this, calculations will be done separately for each species.

```r
stats_table <- iris %>%
     filter(Species %in% c("setosa", "versicolor")) %>%
     group_by(Species) %>%
     summarise(
         mean_length = mean(Sepal.Length),
         median_length = median(Sepal.Length),
         sd_length = sd(Sepal.Length)
     )

stats_table
```

Symbol Breakdown
<-: Assigns values on the right to a variable on the left. Example: setosa <- ... means we’re saving some values in setosa.

$: Accesses a specific column in a dataset.

iris$Sepal.Length selects only the Sepal.Length column from iris.
[ ]: Filters or selects data based on a condition inside.

iris$Species == "setosa" checks where Species is exactly "setosa" and returns only those rows for Sepal.Length.
==: Tests if values are equal.

iris$Species == "setosa" is TRUE where Species is "setosa" and FALSE otherwise.
